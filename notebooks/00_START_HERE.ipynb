## 常见问题

- **`ModuleNotFoundError: llm_course`**：关闭当前服务，回到仓库根目录运行 `uv sync`，再用 `uv run llm-course lab`。
- **Kernel 不是项目环境**：右上角选择 Python 3；仍不正确时关闭 Jupyter，并使用课程命令重新启动。
- **端口被占用**：运行 `uv run llm-course lab --port 8890`。
- **远程机器没有浏览器**：运行 `uv run llm-course lab --no-browser`。
- **不知道下一步**：回到 [15 周核心路径](../docs/core_learning_path.md)，只看当前周，不要同时打开全部论文和源码。
